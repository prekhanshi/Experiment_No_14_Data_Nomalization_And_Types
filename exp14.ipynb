```python
#dataset
import pandas as pd
import numpy as np

data = {
    'product': [ 'laptop','mobile', 'tablet','headphones','smartwatch','camera' ],
    'price' : [55000,20000,30000,2000,8000,45000],
    'units_sold': [25,60,35,80,50,15],
    'discount': [10,15,12,20,18,8]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1.min-max normalization
# min max normalization for price
df['price_MinMax'] = (df['price'] - df['price'].min()) / (df['price'].max() - df['price'].min())
print("\nMin-Max Normalized Data:")
print(df[['product','price','price_MinMax']])
```

```python
#normalize multiple columns
cols = ['price','units_sold','discount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print ('\nNormalized Dataset')
print(df_norm)
```

```python
#2. Z score normalization
df['units_zscore'] = (df['units_sold'] - df['units_sold'].mean()) / df['units_sold'].std()
print("\nZ-Score Normalized units sold:")
print(df[['product','units_sold','units_zscore']])
```

```python
#3. decimal scaling
df['price_decimal'] = df['price']/ 100000
print('\ndecimal scaling')
print(df[['product', 'price', 'price_decimal']])
```

```python
# part B: turning categorical variables into quantitative variables
import pandas as pd
data= {
'Order_ID' : [101,182,283,204,105,106],
'Customer _Gender': ['Male','Female', 'Female', 'Male', 'Female', 'Male'],
"Payment_Method" : ["UPI", 'Credit Card', 'Debit Card', 'COD', 'UPI', 'Debit Card'],
'Product _Category': ['Electronics', 'Clothing', 'Home', 'Electronics', 'Beauty', 'Electronics'],
"City": ['Pune', 'Mumbai', 'Delhi', 'Bangalore', 'Pune', 'Hyderabad'],
"Order _Value" : [4509,2388,3288,5208,1800,4108]
}
df = pd.DataFrame(data)
df
```

```python
# method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender_Label'] = le.fit_transform(df['Customer _Gender'])
df
```

```python
#method 2: one hot encodng
df_encoded = pd.get_dummies(df, columns=['Payment_Method'])
df_encoded
```

```python
#method 3: encoding multiple columns
df_multi = pd.get_dummies(df, columns=['Product _Category','City'])
print(df_multi)
```

```python
#Method 3: dummy encoding (drop first)
df_dummy = pd.get_dummies(df, columns=['Payment_Method'], drop_first=True)
print(df_dummy)
```
